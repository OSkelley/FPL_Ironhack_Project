```python
import pandas as pd
```

```python
pl_season = 2019
```

```python
player_data = pd.read_csv(f'{pl_season}_merge.csv')
```

```python
player_data
```

```python
for x in range(0, (len(player_data)-1)):
    gca = pd.read_csv(f"{pl_season}_stats/{player_data['id'][x]}_gca.csv")
    passing = pd.read_csv(f"{pl_season}_stats/{player_data['id'][x]}_passing.csv")
    summary = pd.read_csv(f"{pl_season}_stats/{player_data['id'][x]}_summary.csv")
    passing_merge = passing.drop(columns= ['Unnamed: 0', 'Date', 'Day', 'Venue', 'Result', 'Squad', 'Opponent', 'Start', 'Pos', 'Min', 'Cmp', 'Att', 'Cmp%', 'Att.1', 'Ast', 'xAG', 'PrgP', 'Match Report'])
    gca_merge = gca.drop(columns= ['Unnamed: 0', 'Date', 'Day', 'Venue', 'Result', 'Squad', 'Opponent', 'Start', 'Pos', 'Min', 'SCA', 'Sh', 'GCA', 'Match Report'])
    summary_merge = summary.drop(columns= ['Unnamed: 0', 'Match Report'])
    merge1 = summary_merge.merge(gca_merge, on = 'Round')
    merge2 = merge1.merge(passing_merge, on= 'Round')
    merge2 = merge2.replace('On matchday squad, but did not play', None)
    merge2 = merge2.dropna()
    merge2.to_csv(f"{pl_season}_stats/{player_data['id'][x]}_merge.csv")
```
