```python
import pandas as pd
```

```python
pl_season = 2017
```

```python
player_data = pd.read_csv(f"../Web-Scraping/{pl_season}_merge.csv")
```

```python
concat = pd.read_csv(f"../Web-Scraping/{pl_season}_stats/{player_data['id'][0]}_merge.csv")
concat['id'] = player_data['id'][0]
for x in range(1, (len(player_data)-1)):
    player = pd.read_csv(f"../Web-Scraping/{pl_season}_stats/{player_data['id'][x]}_merge.csv")
    player['id'] = player_data['id'][x]
    concat = pd.concat([concat, player], ignore_index = True)

# concat = concat.drop(columns = ['Fls', 'Fld_x', 'Off', 'Crs', 'TklW', 'OG', 'PKwon', 'PKcon', 'Fld_y'])
concat = concat.drop(columns= 'Unnamed: 0')
concat.to_csv(f"{pl_season}_all.csv")
```
