```python
import pandas as pd
```

```python
pl_season = 2022
```

```python
data = pd.read_csv(f"{pl_season}_all.csv")
unique_id = pd.read_csv(f"unique{pl_season}.csv")
id_df = unique_id[['id', 'unique']]
```

```python
data
```

```python
join = data.join(id_df, on= 'id', how= 'left', lsuffix='_left')
```

```python
join = join.drop(columns= ['Unnamed: 0', 'id_left'])
join
```

```python
join.to_csv(f"{pl_season}_all.csv")
```
